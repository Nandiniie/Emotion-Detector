```python
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("All tools ready!")
```

```python
import urllib.request
import zipfile
import os

print("Downloading dataset... please wait 2-3 minutes...")

# Download directly from Zenodo — official free source
urllib.request.urlretrieve(
    "https://zenodo.org/record/1188976/files/Audio_Speech_Actors_01-24.zip",
    "ravdess.zip"
)

print("Download done! Now unzipping...")

with zipfile.ZipFile("ravdess.zip", "r") as z:
    z.extractall("ravdess_data")

print("Done! Let us check what we got:")
count = 0
for root, dirs, files in os.walk("ravdess_data"):
    for file in files:
        if file.endswith(".wav"):
            count += 1
print(f"Total audio files found: {count}")
```

```python
emotion_map = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

print("Emotions we will detect:")
for code, emotion in emotion_map.items():
    print(f"  Code {code} = {emotion}")
```

```python
def extract_features(file_path):
    """
    This function reads one audio file
    and converts it into 40 numbers.
    Those 40 numbers describe how the voice sounds.
    Think of it like a voice fingerprint!
    """
    try:
        # Load the audio file
        audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')

        # Extract MFCC — 40 numbers describing the sound
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)

        # Take average of each number across time
        mfccs_mean = np.mean(mfccs.T, axis=0)

        return mfccs_mean

    except Exception as e:
        print(f"Error with file {file_path}: {e}")
        return None

print("Feature extractor ready!")
```

```python
features = []
labels = []

dataset_path = "ravdess_data"

print("Loading all audio files... please wait...")

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)

            # Get emotion from filename
            parts = file.split("-")
            if len(parts) >= 3:
                emotion_code = parts[2]

                if emotion_code in emotion_map:
                    emotion = emotion_map[emotion_code]

                    # Extract the 40 numbers from this file
                    feat = extract_features(file_path)

                    if feat is not None:
                        features.append(feat)
                        labels.append(emotion)

print(f"Total audio files loaded: {len(features)}")
print(f"Emotions found: {set(labels)}")
```

```python
X = np.array(features)
y = np.array(labels)

print("Input shape:", X.shape)
print("Output shape:", y.shape)

# See how many files per emotion
unique, counts = np.unique(y, return_counts=True)
print("\nFiles per emotion:")
for emotion, count in zip(unique, counts):
    print(f"  {emotion}: {count} files")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=y, palette='Set2')
plt.title('How many audio files per emotion?')
plt.xlabel('Emotion')
plt.ylabel('Number of files')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Encode emotions as numbers (computer needs numbers not words)
# happy = 0, sad = 1, angry = 2 etc automatically

le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Emotion to number mapping:")
for i, emotion in enumerate(le.classes_):
    print(f"  {emotion} = {i}")

# Split 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42
)

print(f"\nTraining files: {len(X_train)}")
print(f"Testing files: {len(X_test)}")
```

```python
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("All tools ready!")
```

```python
!pip install gdown

import gdown
import zipfile

# Download RAVDESS dataset (free emotion speech dataset)
url = "https://drive.google.com/uc?id=1wWsrN2ZvAbx1IDuGspZfHHSDTHBSVxFl"
gdown.download(url, "ravdess.zip", quiet=False)

# Unzip it
with zipfile.ZipFile("ravdess.zip", "r") as z:
    z.extractall("ravdess_data")

print("Dataset downloaded and ready!")
```

```python
# RAVDESS dataset has audio files of actors speaking with different emotions
# The emotion is hidden inside the filename!
# Example filename: 03-01-05-01-01-01-01.wav
# The 3rd number tells the emotion:

emotion_map = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

print("Emotions we will detect:")
for code, emotion in emotion_map.items():
    print(f"  Code {code} = {emotion}")
```

```python
def extract_features(file_path):
    """
    This function reads one audio file
    and converts it into 40 numbers.
    Those 40 numbers describe how the voice sounds.
    Think of it like a voice fingerprint!
    """
    try:
        # Load the audio file
        audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')

        # Extract MFCC — 40 numbers describing the sound
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)

        # Take average of each number across time
        mfccs_mean = np.mean(mfccs.T, axis=0)

        return mfccs_mean

    except Exception as e:
        print(f"Error with file {file_path}: {e}")
        return None

print("Feature extractor ready!")
```

```python
features = []
labels = []

dataset_path = "ravdess_data"

print("Loading all audio files... please wait...")

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)

            # Get emotion from filename
            parts = file.split("-")
            if len(parts) >= 3:
                emotion_code = parts[2]

                if emotion_code in emotion_map:
                    emotion = emotion_map[emotion_code]

                    # Extract the 40 numbers from this file
                    feat = extract_features(file_path)

                    if feat is not None:
                        features.append(feat)
                        labels.append(emotion)

print(f"Total audio files loaded: {len(features)}")
print(f"Emotions found: {set(labels)}")
```

```python
X = np.array(features)
y = np.array(labels)

print("Input shape:", X.shape)
print("Output shape:", y.shape)

# See how many files per emotion
unique, counts = np.unique(y, return_counts=True)
print("\nFiles per emotion:")
for emotion, count in zip(unique, counts):
    print(f"  {emotion}: {count} files")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=y, palette='Set2')
plt.title('How many audio files per emotion?')
plt.xlabel('Emotion')
plt.ylabel('Number of files')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Encode emotions as numbers (computer needs numbers not words)
# happy = 0, sad = 1, angry = 2 etc automatically

le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Emotion to number mapping:")
for i, emotion in enumerate(le.classes_):
    print(f"  {emotion} = {i}")

# Split 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42
)

print(f"\nTraining files: {len(X_train)}")
print(f"Testing files: {len(X_test)}")
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("Scaling done!")
```

```python
print("Training the model... please wait...")

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

print("Model trained! Computer has learned emotions!")
```

```python
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"Accuracy: {accuracy * 100:.2f}%")
print()
print("Full report:")
print(classification_report(y_test, predictions,
      target_names=le.classes_))
```

```python
cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.title('Confusion Matrix — Emotion Recognition')
plt.xlabel('What computer predicted')
plt.ylabel('What was actually true')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('emotion_confusion_matrix.png')
plt.show()
```
